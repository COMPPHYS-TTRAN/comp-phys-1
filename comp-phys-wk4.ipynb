#1. arrays and numpy

-all elements in an array must be of the same type, usually integers, real or complex numbers, or strings for efficient numerical computing and storage
- the number of elements must be known when the array is created
-arrays are not part of the standard python- one needs an additional package called Numerical Python: NumPy.
-a wide range of mathematical operations can be done directly on complete arrays, thereby removing the need for loops over array elements. This is commonly called vectorization and may cause a dramatic speed-up of python programs. 
- arrays with one index are often called vectors. Arrays with two indices are used as an efficient data structure for tables, instead of lists of lists. Arrays can also have three or more indices.

```python
#temp conv using arrays

def temp_convert(tstart, tend, numsteps):
    c_temp = np.linspace(tstart, tend, numsteps)
    f_temp = 9.0/5.0*(c_temp)+32.
    return f_temp

print temp_convert( 0, 100, 2)
```
